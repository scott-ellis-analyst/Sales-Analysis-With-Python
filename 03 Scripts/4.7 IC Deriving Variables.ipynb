## 2. Applying price label on subset

```python
# creating a subset of the dataframe 

df = df_ords_prods_merge[:1000000]
```

```python
# checking dimensions

df.shape
```

```python
# creating price label function

def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# applying price_labe function on df

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# checking value counts

df['price_range'].value_counts(dropna = False)
```

```python
# checking max price in subset

df['prices'].max()
```

```python
# checking output

df.head()
```

```python
# next 3 cells are using loc to create the new column

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# checking value counts

df['price_range_loc'].value_counts(dropna = False)
```

```python
# checking output

df.head()
```

## 4. Creating 'busiest_day' column

```python
# checking order value counts for days of the week

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# create for loop labelling each order in terms of busyness

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# checking output of result list from for loop

result
```

```python
# creating new column in dataframe and populating it with result list

df_ords_prods_merge['busiest_day'] = result
```

```python
# checking value counts

df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
# checking output

df_ords_prods_merge.head()
```

## 5. Changing 'busiest_day' to 'busiest_days'

```python
# checking order value counts for days of the week

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

### Top 2 busiest days are 0 = Saturday and 1 = Sunday. 2 slowest days are 3 = Tuesday and 4 = Wednesday.

```python
# create for loop labelling each order in terms of busyness

result2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result2.append("Busiest days")
  elif value == 4 or value == 3:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
# creating new column in dataframe and populating it with result2 list

df_ords_prods_merge['busiest_days'] = result2
```

```python
# checking value counts

df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

### 'busiest_days' value counts match up with 'orders_day_of_week'

```python
# checking output

df_ords_prods_merge.head()
```

## 6. Creating 'busiest_period_of_day' column

```python
# checking order value counts for hours of day

df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

### Hour categories: 0-6, 23 = Fewest orders / 7,8, 17-22 = Average orders / 9-17 = Most orders

```python
# create for loop labelling each order in terms of hour category

result_hour = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [9,10,11,12,13,14,15,16,17]:
    result_hour.append("Most orders")
  elif value in [0,1,2,3,4,5,6,23]:
    result_hour.append("Fewest orders")
  else:
    result_hour.append("Average orders")
```

```python
# creating new column in dataframe and populating it with result_hour list

df_ords_prods_merge['busiest_period_of_day'] = result_hour
```

```python
# checking order value counts for busiest_period_of_day

df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
